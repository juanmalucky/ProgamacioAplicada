```python
# Create dataframe from a list

import pandas as pd
```

```python
numbers = [1, 2, 3, 4, 5, 6]# Se crea un DataFrame a partir de una lista de números
```

```python
pd.DataFrame(numbers)
```

```python
data = [["Jhon",  23], ["Mary", 18], ["Jhon", 31]]# Se crea un DataFrame a partir de una lista de listas (datos con nombre y edad)
```

```python
pd.DataFrame(data)
```

```python
pd.DataFrame(data, columns = ["name", "age"])
```

```python
pd.DataFrame(data, columns = ["name", "age"], index = ["id1", "id2", "id3"])# Con nombres de columnas e índices
```

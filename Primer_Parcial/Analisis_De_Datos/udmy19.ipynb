```python
#Refine Dataframe using Multiple Conditions AND

import pandas as pd

#insta = pd.read_csv('/content/drive/MyDrive/udmy/dataAnalisys/Instagram.csv')
insta = pd.read_csv('/content/drive/MyDrive/udmy/dataAnalisys/Instagram.csv', parse_dates = ["reg date", "last login"])
insta["gender"] = insta["gender"].astype("category")
insta["Country"] = insta["Country"].astype("category")
insta["Influencer"] = insta["Influencer"].astype("bool")

insta.head()
```

```python
#Print our Male instagram users that joined since 2016 and they are from USA

mask = (insta["gender"] == "Male") & (insta["reg date"] > "2015-12-31") & (insta["Country"] == "United States")# Crea una máscara para filtrar usuarios hombres de USA que se registraron después de 2015
insta[mask]# Aplica la máscara para obtener esos usuarios
```

```python
mask1 = insta["gender"] == "Male"
mask2 = insta["reg date"] > "2015-12-31"
mask3 = insta["Country"] == "United States"
insta[mask1 & mask2 & mask3]# Lo mismo pero separando las condiciones en variables para mayor claridad
```

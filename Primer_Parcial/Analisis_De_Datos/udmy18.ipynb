```python
import pandas as pd

insta = pd.read_csv('/content/drive/MyDrive/udmy/dataAnalisys/Instagram.csv')
insta["reg date"] = pd.to_datetime(insta["reg date"])
insta["last login"] = pd.to_datetime(insta["last login"])
insta["gender"] = insta["gender"].astype("category")
insta["Country"] = insta["Country"].astype("category")
insta["Influencer"] = insta["Influencer"].astype("bool")
insta.info()
```

```python
import pandas as pd

insta = pd.read_csv('/content/drive/MyDrive/udmy/dataAnalisys/Instagram.csv', parse_dates = ["reg date", "last login"])# Cambia los tipos de datos de algunas columnas para optimizar memoria y facilitar análisis
insta["gender"] = insta["gender"].astype("category")
insta["Country"] = insta["Country"].astype("category")
insta["Influencer"] = insta["Influencer"].astype("bool")
```
